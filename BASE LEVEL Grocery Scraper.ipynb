```python
# Recipe input / user input
ingredients = ['taco shells' ,'ground beef', 'shredded cheese mexican blend', 'avocado', 'taco sauce']
```

```python
products_main_list = []

for ing in ingredients:
    print(ing)

    input_field = browser.find_by_id('typeahead-search-input')
    input_field.fill(ing)
    search_button = browser.find_by_css('button.search-button')
    search_button.click()

    time.sleep(10)

    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')

    product_list = soup.find('ul', class_='product-list')
    products = product_list.find_all('li')

    for p in products:
        try:
            price = p.find('span', class_='product-grid-cell_main-price').text
            title = p.find('div', class_='product-tile_detail-title').h3.text
            size = p.find('span', class_='product-grid-cell_size').text
            unit = p.find('span', class_='product-grid-cell_unit').text

            product_dict = {
                'ingredient': ing,
                'price': price,
                'title': title,
                'size': size,
                'unit': unit
            }

            products_main_list.append(product_dict)

        except Exception as e:
            print(e)
            pass
```

```python
product_df = pd.DataFrame(products_main_list)
pd.set_option('display.max_rows', None)
product_df
```
